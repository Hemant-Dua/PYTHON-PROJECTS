```python
import pandas as pd
import matplotlib as plt

num=int(input("Enter your number: "))
iter=0
ind=[]
number=[]

while(num!=1):
    iter+=1

    if (num%2==0):
        num=num/2
    else:
        num=(3*num)+1

    ind.append(iter)
    number.append(int(num))
    
df=pd.DataFrame({'Values':number},index=ind)
```

```python
df.plot(kind='line', 
		use_index=True, 
		y='Values', 
		color='red') 
```
